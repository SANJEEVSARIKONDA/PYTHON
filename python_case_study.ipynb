```python
pd.set_option('display.max_columns',None)
```

```python
df_sale.head()
```

1. Creat new flag variable Late Delivery Risk based on Days for shipping (real) & Days for shipment (scheduled) (Flag=Not Late if shipment is not delayed and Flag=Late if shipment is delayed)

```python

df_sale['late_delivery']=df_sale['shippingreal_days']-df_sale['shipmentscheduled_day']

```

```python
df_sale['risk']=np.where(df_sale['late_delivery']>0 ,'delayed','not delayed')
```

```python
df_sale
```

2. Rename the variables as per python naming convensions

```python
df_sale.rename(columns={'order date (DateOrders)':'Orders_Date ','shipping date (DateOrders)':'shipping_date','Product Category Id':'Category Id','Days for shipment (scheduled)':'shipmentscheduled_day','Days for shipping (real)':'shippingreal_days'},inplace= True)
```

3. Convert the variables data types as per the data descriptions

```python

df_sale['shipping_date']= pd.to_datetime(df_sale['shipping_date'])
df_sale['Orders_Date ']= pd.to_datetime(df_sale['Orders_Date '])
```

```python
df_sale.isnull().sum()
```

4. Impute missing values with mean for numerical variables, mode for categorical variables

```python
df_sale.dropna(axis=0,how='any',inplace=True)
```
